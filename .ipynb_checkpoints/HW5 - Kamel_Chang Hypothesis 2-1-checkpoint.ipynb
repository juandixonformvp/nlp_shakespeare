### In this analysis, we will perform cluster analysis on the characters in Shakespeare plays. Scholars have done analysis previously on the similarity of many of Shakespeare's characters, for example, the similarity of Macbeth and Hamlet. We will take the main 5 characters from each of Shakespeare's 37 plays (37 x 5 = 185) and use natural language processing to find clusters. We use two clustering techniques. 1) K-means, using a bag of words approach, and 2) Spectral Clustering, using sentiment analysis through Word2Vec.

```python
VERBOSE = True
BASE_URL = 'http://shakespeare.mit.edu/'

# Get main page
r = requests.get(BASE_URL)
assert(r.status_code == 200)

# Soupify main page and get table element
soup = BeautifulSoup(r.content, 'html.parser')
table = soup.find_all('table')[1]

# Scrape play titles from table element
titles = {}  # dict of plays and play data
num_titles = 0
if not VERBOSE:
    print('Getting documents', end='')
start = time.time()
for link in table.find_all('a'):
    # Skip poetry
    if 'Poetry' not in link.get('href'):
        if '\n' in link.string:
            link.string = link.string.replace('\n', ' ').strip()
        href = link.get('href').replace('index', 'full')
        # Taking advantage of their URL naming conventions
        send_url = BASE_URL + href
        if VERBOSE:
            print('Getting {} from {}'.format(link.string, send_url))
        else:
            print('.', end='')
        num_titles += 1
        # Get text of each play and add to dict of plays
        r = requests.get(send_url)
        assert(r.status_code == 200)
        # `titles` dict structure:
        titles[link.string] = {
            'href': href,
            'soup': BeautifulSoup(r.content, 'html.parser')
        }
end = time.time()
print('Done! Retrieved {} documents in {:.3f}s.'.format(num_titles, end - start))
```

### We scrape the Shakespeare text from the MIT website. Full credit for the web scraping code is attributed to Mayoya Masaki's Github page. In the other hypothesis of sentiment analysis, we have based the code on section and updated to download everything using beautifullsoup

```python
COLUMN_NAMES = ['Title','Act','Scene','Speaker','Words']


# Given the data in the parameters, decides how to add the data to the data frame.
def add_data(df, title, act, scene, speaker, words):
    # Do we already have an entry for this speaker?
    if speaker in df.Speaker.values:
        # Get the row index of this speaker and add the new words to that speaker's existing words
        indices = list(df[df.Title == title][df.Act == act][df.Scene == scene][df.Speaker == speaker].index)
        if len(indices) == 1:
            index = indices[0]
            df.set_value(index, 'Words', df.Words[index] + ' ' + words)
            return df
        # In case of weird errors, but there should never be more than one row fitting above criteria
        elif len(indices) > 1:
            print(title, act, scene, speaker, indices)
            return df
    # If this is a totally new speaker, make a new row entry in the data frame for him/her
    new_row = pd.DataFrame([[title, act, scene, speaker, words]], columns=COLUMN_NAMES)
    df = df.append(new_row, ignore_index=True)
    return df


data = pd.DataFrame([], columns=COLUMN_NAMES)
num_processed = 0
if not VERBOSE:
    print('Processing documents', end='')

# For each play, iterate through each line of the play's html text sequentially,
#  tracking which act, scene, and speaker we're on. Whenever a new speaker
#  or a new scene/act arrives, take the words we've gathered for the last speaker
#  and add them to the data frame.
# The resulting table will contain 1 row for each unique combination of play,
#  act, scene, and speaker. Thus the words for each speaker are grouped by scene.
#        ** NB: This script ONLY scrapes character speeches - if
#        **     prologue material exists without a speaker specified
#        **     (i.e., "narrator" text), this script will ignore it.
start = time.time()
for title in titles:
    if VERBOSE:
        print(title)
    else:
        print('.', end='')
    # Skip to the first act - which is not always ACT I but sometimes PROLOGUE or even None
    act = titles[title]['soup'].find('h3')
    scene = speaker = words = ''
    # The rest of this is essentially just a big nested switch statement.
    if act.string[:3] != 'ACT':
        scene = act.string.split('.')[0]
        act.string = 'None'
    for elem in act.next_elements:
        if elem.name == 'h3':
            if words and speaker:
                data = add_data(data, title, act.string, scene, speaker, words)
                words = ''
            if elem.string[:3] == 'ACT':
                act = elem
            else:
                scene = elem.string.split('.')[0]
        if elem.name == 'a':
            if elem.attrs['name'][:6] == 'speech':
                if words and speaker:
                    data = add_data(data, title, act.string, scene, speaker, words)
                    words = ''
                speaker = elem.string.lower()
            else:
                if words and speaker:
                    words += ' ' + elem.string.strip()
                else:
                    words += elem.string.strip()
    num_processed += 1
end = time.time()
print('Done! Processed {} documents in {:.3f}s.'.format(num_processed, end - start))
```

### We only want to cluster Shakespeare's main characters, not the minor characters. We take the top 5 characters in each Shakespeare play, who speak the most words.

```python
dfSpeak = data.groupby(['Title','Speaker']).Words.apply(sum).reset_index()
dfSpeak['chunkLen'] = dfSpeak.apply(lambda row: len(row.Words), axis=1)
dfSpeak.head(10)
```

### Our dataframe now contains the title, speaker, all words spoken by that speaker, and the word count. For example we see that from A Midsummer Night's Dream, our top 5 characters are bottom, oberon, helena, lysander, and theseus.

```python
dfTopSpeak = dfSpeak.sort_values(['chunkLen'],ascending=False).groupby('Title').head(5)
dfTopSpeak = dfTopSpeak.sort_values(['Title'])
dfTopSpeak = dfTopSpeak.reset_index(drop=True)
titles = dfTopSpeak.Title
speaker = dfTopSpeak.Speaker
words_chunk = dfTopSpeak.Words
dfTopSpeak.head(10)
```

### We need to designate stop words to filter out, and decide on a stemmer. We experimented with several tokenizers, with the idea being that we wanted to remove most punctuation and white space, and eliminate fragments such as 's.

```python
# load nltk's English stopwords as variable called 'stopwords'
all_stopwords = stopwords.words('english')
#all_stopwords.append('\'') 
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
from nltk.tokenize import TweetTokenizer

def tokenize_and_stem(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems


def tokenize_only(text):
    # first tokenize by sentence, then by word to ensure that punctuation is caught as it's own token
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = []
    # filter out any tokens not containing letters (e.g., numeric tokens, raw punctuation)
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
    return filtered_tokens
```

### I create a pandas DataFrame with the stemmed vocabulary as the index and the tokenized words as the column. The benefit of this is it provides an efficient way to look up a stem and return a full token

```python
totalvocab_stemmed = []
totalvocab_tokenized = []
for i in words_chunk:
    allwords_stemmed = tokenize_and_stem(i) #for each item in 'synopses', tokenize/stem
    totalvocab_stemmed.extend(allwords_stemmed) #extend the 'totalvocab_stemmed' list
    
    allwords_tokenized = tokenize_only(i)
    totalvocab_tokenized.extend(allwords_tokenized)
    
vocab_frame = pd.DataFrame({'vocab': totalvocab_tokenized}, index = totalvocab_stemmed)
print('there are ' + str(vocab_frame.shape[0]) + ' items in vocab_frame')
print(vocab_frame.head(50))
```

### We create a TFIDF matrix to use for clustering. Brandon Rose's "brandonrose.org/clustering" page was used as a reference.

```python
#define vectorizer parameters
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, min_df=0.2, max_features=1000,
                                 stop_words=all_stopwords,
                                 use_idf=True, tokenizer=tokenize_and_stem, ngram_range=(1,1))

%time tfidf_matrix = tfidf_vectorizer.fit_transform(words_chunk) #fit the vectorizer to synopses

print(tfidf_matrix.shape)

vocab = tfidf_vectorizer.get_feature_names()
tfidf_matrix = tfidf_matrix.A

df = pd.DataFrame(data=tfidf_matrix,columns=vocab)
print(df.shape)
df.head()
```

### We choose K=5 for K-means. With K=5 the K-means is slightly unstable, but we mostly get one large cluster of of 80 characters and 4 roughly equal sized clusters containing 20-30 characters.

```python
num_clusters = 5
km = KMeans(n_clusters=num_clusters)
%time km.fit(tfidf_matrix)
clusters = km.labels_.tolist()

speakDict = { 'Title': titles, 'Speaker': speaker, 'cluster': clusters }
frame = pd.DataFrame(speakDict , columns = ['Title', 'Speaker', 'cluster'])
frame['cluster'].value_counts()
```

### We can check if the clusters make intuitive sense. We print each cluster, along with the "top terms" associated with each cluster. We can see that the clustering is roughly separating Shakespeare's historical characters, from his tragic characters, from his comic characters. Common terms from the historical characters include terms such as "king" and "queen", while the comic characters are associated with terms such as "fool" and "sweet".

```python
grouped = frame['Title'].groupby(frame['cluster'])

#from __future__ import print_function
print("Top terms per cluster:")
print()
#sort cluster centers by proximity to centroid
order_centroids = km.cluster_centers_.argsort()[:, ::-1] 

for i in range(num_clusters):
    print("Cluster %d words:" % i, end='')
    
    for ind in order_centroids[i, :6]: #replace 6 with n words per cluster
        print(' %s' % vocab_frame.ix[vocab[ind].split(' ')].values.tolist()[0][0].encode('utf-8', 'ignore'), end=',')
    print() #add whitespace
    print() #add whitespace
    
    print("Cluster %d speakers:" % i, end='')
    print() #add whitespace
    print(frame['Speaker'].loc[frame['cluster'] == i])
    print() #add whitespace
    print() #add whitespace
    
print()
print()
```

### Once we've generated the model, we can perform various NLP word tasks. We use the "n_similarity" function of gensim to test similarity between our 185 Shakespeare characters. Gensim takes the mean vector of each set of vector, and then computes the cosine similarity of the resultant vector means.

```python
    def w2v(s1,s2,model):
            if s1==s2:
                return 1.0

            s1words=s1.split()
            s2words=s2.split()
            
            model_vocab = model.wv.vocab #the vocabulary considered in the word embeddings
                    
            for word in s1words: #remove sentence words not found in the vocab
                if (word not in model_vocab):
                    #print(word)
                    s1words.remove(word)
            
            for word in s2words: #remove sentence words not found in the vocab
                if (word not in model_vocab):
                    #print(word)
                    s2words.remove(word)
            
            return model.n_similarity(s1words, s2words)
```
